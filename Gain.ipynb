### ABS 

```python
import pandas as pd 
import matplotlib.pyplot as plt
def Gain(zero,faut):
    return (zero*10000-100000*faut)    
```

```python
# Création d'un DataFrame avec les performances
d = {
    'Modèle': ['Régression Logistique', 'RandomForest_weighted','RandomForest', 'XGBoost'],
    'Gain de la Banque ': [Gain(759,20), Gain(757,15), Gain(757,15),Gain(758,16)],
    
}

df = pd.DataFrame(d)

fig, ax = plt.subplots(figsize=(10,2))  
ax.axis('tight')
ax.axis('off')


table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)  
table.scale(1.5, 1.5) 

plt.title("Gain de la Banque par Rapport à Chaque Modèle  ABS ")
plt.show()
```
